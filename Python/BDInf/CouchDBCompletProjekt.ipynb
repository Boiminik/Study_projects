# Apache CouchDB

## Special Features and Concepts:

Document Store: CouchDB stores data in a schema-less manner, where each record is a document containing key-value pairs.
Distributed Architecture: CouchDB employs a distributed architecture, enabling replication and synchronization of data across multiple nodes.
HTTP API: It provides a RESTful HTTP API for accessing and manipulating data, making it accessible from any programming language or platform.
MapReduce Views: CouchDB allows users to create views using MapReduce functions, facilitating efficient querying and aggregation of data.


## Documentation:

Documentation for CouchDB can be found on its official website
https://docs.couchdb.org/en/stable/

## Typical Application Areas:

Collaborative Applications: CouchDB is suitable for applications requiring distributed data storage and synchronization, such as collaborative document editing tools.
Offline-First Applications: It's ideal for applications that need to work offline and sync data with a central database when online.
Content Management Systems (CMS): CouchDB can be used as a backend for CMS platforms due to its flexible document storage model.

## STRENGTHS:

Flexible Schema: CouchDB's schema-less design allows for easy adaptation to evolving data requirements.
Replication and Synchronization: Its distributed nature enables efficient replication and synchronization of data across multiple nodes.
Offline Support: CouchDB's architecture supports offline-first applications, making it resilient to network failures.

## WEAKNESSES:

Performance: While CouchDB excels in distributed scenarios, it may not offer the same level of performance as some other databases for certain use cases.
Learning Curve: Users unfamiliar with document-oriented databases may face a learning curve when transitioning to CouchDB.
Lack of Full ACID Compliance: CouchDB sacrifices some aspects of ACID compliance for the sake of scalability and flexibility.

# Connect to CouchDB

URL to the CouchDB instance with authentication information

```python
url = 'http://admin:password@localhost:5984/'
```

Establish connection to the CouchDB server

```python
couch = couchdb.Server(url)
```

Name of the new database

```python
db_name = 'example_db'
```

Connect to the CouchDB server

```python
couch = couchdb.Server(url)
```

Check if the database already exists

```python
if db_name in couch:
    print(f'Database "{db_name}" already exists.')
    db = couch[db_name]  # Auf eine vorhandene Datenbank zugreifen
else:
    # Neue Datenbank mit dem angegebenen Namen erstellen
    db = couch.create(db_name)
    print(f'Created new database "{db_name}".')
```

## Insert document into the database

Generate a unique ID manually

```python
doc_id = str(uuid.uuid4())  # UUID generieren und in einen String konvertieren
doc = {
    '_id': doc_id,
    'name': 'Johnny Doe',
    'age': 31
}
```

Add document to the database

```python
db.save(doc)
print(f'New document added with ID "{doc_id}".')
```

Output that the connection has been establishe

```python
print(f'Connected to database: {db}')
```

## Read from the DB

Retrieve all documents in the database

```python
all_docs = db.view('_all_docs', include_docs=True)
```

Print all documents

```python
for row in all_docs:
    doc = row['doc']
    print(doc)
```

#  Query language

```python
# Retrieve design document if it exists
design_doc_id = '_design/mydesign'
if design_doc_id in db:
    print(f'Design document "{design_doc_id}" already exists.')
    design_doc = db[design_doc_id]
else:
    # Map function defined in JavaScript
    map_function = '''
    function(doc) {
        if (doc.type == 'person') {
            emit(doc.name, 1);
        }
    }
    '''

    # Reduce function defined in JavaScript
    reduce_function = '''
    function(keys, values, rereduce) {
        return sum(values);
    }
    '''

    # Design document created and sent to the database
    design_doc = {
        '_id': '_design/mydesign',
        'views': {
            'myview': {
                'map': map_function,
                'reduce': reduce_function
            }
        }
    }
    db.save(design_doc)
    print(f'Design document "{design_doc_id}" created and saved.')

# Query with MapReduce
result = db.view('mydesign/myview')
print(result)
```

Delete the new Database

```python
import requests

session = requests.Session()
# Admin username and password from Fauxton
session.auth = ('admin', 'password')
sesscode = session.delete(f"http://127.0.0.1:5984/delete_me") # URL of database to delete

if sesscode.status_code == 200: # Message if ok
    print(f"Database deleted successfully.")
elif sesscode.status_code == 404: # Message if name unknown
    print(f"Database not found.")
else: # else
    print(f"Failed to delete database. Status code: {sesscode.status_code}")
```

Import database and create container.

```python
import docker

def create_couchdb_container():
    # Initialize Docker client
    client = docker.from_env()

    # Pull the CouchDB image
    client.images.pull('couchdb')

    # Define container configuration
    container_config = {
        'image': 'couchdb',
        'name': 'example_container',
        'ports': {'5984/tcp': 5984},  # Expose CouchDB port
        'detach': True  # Run container in detached mode
    }

    # Create the CouchDB container
    container = client.containers.run(**container_config)

    # Print container ID
    print(f"CouchDB container created with ID: {container.id}")

if __name__ == "__main__":
    create_couchdb_container()
```

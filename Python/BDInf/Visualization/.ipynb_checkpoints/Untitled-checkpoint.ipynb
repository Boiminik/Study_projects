# Visualization

## Import Libraries

First we will need to import all of the Libraries that are neccessary:

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Now we need to import the candy.csv file as provided

```python
df = pd.read_csv('candy.csv')
```

## Basic display

Now we display the data in its simplest form:

```python
print(df.head())  # Display first few rows of the DataFrame
print(df.info())  # Get information about the DataFrame

# Check for missing values
print(df.isnull().sum())

# Basic statistical summary
print(df.describe())
```

## Visualization

Now lets make some graphs:

### Distribution of sugar percent

The following graph shows the distribution of the sugar percent for all of the candies:

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='sugarpercent', bins=20, kde=True)
plt.title('Distribution of Sugar Percent')
plt.xlabel('Sugar Percent')
plt.ylabel('Frequency')
plt.show()
```

### Distribution of price percent

Since we now know that the sugar percent of the candies are all around, with the highest amount being at around 60%, we look at the distribution of the price percent for all of the candies:

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='pricepercent', bins=20, kde=True)
plt.title('Distribution of Price Percent')
plt.xlabel('Price Percent')
plt.ylabel('Frequency')
plt.show()
```

### Bar plot of Chocolate vs Non-Chocolate Candies

Are most candies made from chocolate? With the data provided we can create a Bar plot:

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='chocolate')
plt.title('Chocolate vs. Non-Chocolate Candies')
plt.xlabel('Chocolate')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=['No', 'Yes'])
plt.show()
```

### Pair plot for sugar percent, price percent and win percent

```python
sns.pairplot(df[['sugarpercent', 'pricepercent', 'winpercent']])
plt.suptitle('Pairplot of Sugar Percent, Price Percent, and Win Percent', y=1.02)
plt.show()
```

### Boxplot of win percent by chocolate attribute

Are chocolate candies the best candy around? Lets have a look at the data:

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='chocolate', y='winpercent')
plt.title('Win Percent by Chocolate Attribute')
plt.xlabel('Chocolate')
plt.ylabel('Win Percent')
plt.xticks(ticks=[0, 1], labels=['No', 'Yes'])
plt.show()
```

#### Zeile einer CSV Datei als Listen einlesen

```python
# open file as csvfile, encoding UTF-8
with open(path, encoding='utf8') as csvfile:
    # read lines in lists
    CSVrows = csv.reader(csvfile, delimiter=',')
    
    # print each list
    for row in CSVrows:
        print(row)
```

#### Variante 2

```python
# open file as csvfile, encoding UTF-8
csvfile = open(path, encoding='utf8')

# open file as csvfile, encoding UTF-8,
CSVrows = csv.reader(csvfile, delimiter=',')

# read lines in lists
for row in CSVrows:
        print(row)

# close file
csvfile.close()
```

### Datensätze aus CSV in Liste speichern

```python
# open file as csvfile, encoding UTF-8
csvfile = open(path, encoding='utf8')

# empty list
rows = []

# create csv reader object
CSVrows = csv.reader(csvfile, delimiter=',')

# append rows from reader object to list rows
for row in CSVrows:
    rows.append(row)

# close file
csvfile.close()
```

```python
print(rows)
```

#### print first row from rows

```python
print(rows[0])
```

#### print value of second column from first list

```python
print(rows[0][1])
```

#### print all values from second columns, except first

```python
for row in rows[1:]: # [1:] --> from second entry to end
    print(row[1])
```

### write data to seperate lists

```python
# create empty lists
vornamen = []
nachnamen = []
geschlechter = []
plzs = []
orte = []
strassen = []
hausnummern = []


# append values to lists
for row in rows[1:]:
    vornamen.append(row[0])
    
for row in rows[1:]:
    nachnamen.append(row[1])
    
for row in rows[1:]:
    geschlechter.append(row[2])
    
for row in rows[1:]:
    plzs.append(row[3])
    
for row in rows[1:]:
    orte.append(row[4])
    
for row in rows[1:]:
    strassen.append(row[5])
    
for row in rows[1:]:
    hausnummern.append(row[6])
```

```python
# output list vornamen
print(vornamen)
```

```python
# output first 10 entries from list nachnamen
print(nachnamen[0:10])
```

### Number of entries per list

```python
# read number of entries (len()) and output
print('Einträge in Vornamen: ', len(vornamen))
print('Einträge in Nachnamen: ', len(nachnamen))
print('Einträge in Geschlechter: ', len(geschlechter))
print('Einträge in Postleitzahlen: ', len(plzs))
print('Einträge in Orte: ', len(orte))
print('Einträge in Strassen: ', len(strassen))
print('Einträge in Hausnummern: ', len(hausnummern))
```

### Print number of entries for specific value

```python
print(vornamen.count('Martin'))
```

### Read and print out number of distinct values for each list

```python
# write distinct values to new lists
vornamen_dist = set(vornamen)
nachnamen_dist = set(nachnamen)
geschlechter_dist = set(geschlechter)
plzs_dist = set(plzs)
orte_dist = set(orte)
strassen_dist = set(strassen)
hausnummern_dist = set(hausnummern)

# prinut out number of distinct values per list
print('Anzahl der unterschiedliche Vornamen: ', len(vornamen_dist))
print('Anzahl der unterschiedliche Nachnamen: ', len(nachnamen_dist))
print('Anzahl der unterschiedliche Geschlechter: ', len(geschlechter_dist))
print('Anzahl der unterschiedliche Postleitzahlen: ', len(plzs_dist))
print('Anzahl der unterschiedliche Orte: ', len(orte_dist))
print('Anzahl der unterschiedliche Strassen: ', len(strassen_dist))
print('Anzahl der unterschiedliche Hausnummern: ', len(hausnummern_dist))
```

### Print out vornamen and number of occurrence

```python
for vorname in vornamen_dist:
    print(vorname, ': ', vornamen.count(vorname))
```

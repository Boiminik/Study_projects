# CSV
CSV steht für Comma-separated-values und ist ein Dateiformat. Die Endung des Dateinamens lautet `.csv`.
CSV Dateien werden häufig verwendet um Daten zwischen verschiedenen Programmen auszutauschen:
- Tabellenprogramme
- Statistikprogramme
- Datenbanken

Python stellt ein eigenes Modul zum Arbeiten mit CSV zur Verfügung. 

```python
import csv
```

Das `csv module` ist in der Standardbibliothek und muss nicht extra installiert werden

## Aufbau einer CSV Datei
Klassische CSV Dateien, verwenden ein **comma ,** um einzelne Dateneinträge zu trennen. 
Im europäischen Raum wird jedoch meistens der **Semikolon ;** als Trennzeichen verwendet. 
CSV Datei stellen Tabellendaten dar und man spricht daher auch von Zeilen und Spalten 

Eine csv könnte folgendermaßen aussehen

erster Eintrag, 2. Eintrag, noch ein Eintrag

zweite Zeile, zweite Spalte, dritte Spalte

dritte Zeile, 8, 9

## Parsing / Einlesen

Zum Einlesen von CSV Dateien muss ein `csv.reader` erstellt werden. Zum Erstellen dieses Readers wird ein Filestream Handler benötigt. 

```python
f = open("names.csv")  # Öffnen Filestream und Erzeugen des Handlers 
reader = csv.reader(f) # Erstellen des Readers
print(reader)
f.close()
```

**Einlesen als Liste**

```python
with open("names.csv") as f:
    reader = csv.reader(f) 
    data = list(reader)
    print(data)
```

**Zeilenweise verarbeiten**

```python
with open("names.csv") as f:
    reader = csv.reader(f) 
    for line in reader:
        print(line)
```

### Beispiel

```python
# Erstellen Sie händisch ein CSV-File mit 3 Reihen und 3 Spalten und lesen Sie das File ein 
```

```python
with open("names.csv") as f:
    reader = csv.reader(f) 
    for line in reader:
        print(line)
```

```python
# Geben Sie nur die Einträge in der ersten Spalte aus
```

```python
with open("names.csv") as f:
    reader = csv.reader(f) 
    for line in reader:
        print(line[0])
```

```python
# Geben Sie nur jede zweite Zeile aus
```

```python
with open("names.csv") as f:
    reader = csv.reader(f)
    i=0
    for line in reader:
        if(i%2==0):
            print(line)
        i+=1
```

### Header
CSV Dateien können auch einen Header besitzen. Der Header ist die Erste Zeile und definiert für jede Spalte einen Namen

|Vorname | Nachname | Stadt|
|--------|----------|------|
|Max| Mayer| Linz|
|Saya| Delitz| Graz|
|Moritz| Huber| Innsbruck|
|Sarah| Luimböck| Eisenstadt|


### Beispiel

```python
# Erweitern Sie die von Ihnen erstellte CSV-Datei um einen Header und lesen Sie die Datei ein
```

Zum Einlesen von Dateien mit Header eignet es sich die Daten nicht als eine Liste von Listen zu Speichern sondern als ein Liste von Dictionaries. Hierfür gibt es den `csv.DictReader`

```python
with open("names_mit_header.csv") as f:
    reader = csv.DictReader(f)
    print(reader.fieldnames) # Liefert die Namen der Spalten
```

**Zeilenweise Verarbeiten**

```python
with open("names_mit_header.csv") as f:
    reader = csv.DictReader(f)
    for line in reader:
        print(line)
```

**Zeilenweise Verarbeiten einzelner Spalten über Header**

```python
with open("names_mit_header.csv") as f:
    reader = csv.DictReader(f)
    for line in reader:
        print(line["Vorname"])
```

### Andere Trennzeichen
Sollte die CSV-Datei nicht mittels **comma ,** getrennt sein, kann an ein anderer Seperator/Delimiter beim erstellen des Readers übergeben werden

### Beispiel

```python
# Ersetzen Sie das Trennzeichen in der von Ihnen erstellten CSV-Datei und lesen Sie die Datei ein
```

```python
with open("names_tab.csv") as f:
    reader = csv.reader(f, delimiter="\t") 
    for line in reader:
        print(line)
```

## Schreiben
Zum schreiben von CSV Dateien wird ein `csv.writer` benötigt, welcher wieder einen Filehandler braucht. Wichtig, das File muss im **Write-Modus** geöffnet werden

```python
with open("out.csv", mode="w") as f: 
    writer = csv.writer(f)
```

Zum Schreiben wird die Methode **writerow** verwendet

```python
with open("out.csv", mode="w", newline='') as f: 
    writer = csv.writer(f)
    writer.writerow(["Eine", 1, "A", "Liste"])
```

Ausgabe der erstellten Datei

```python
with open("out.csv") as f:
    reader = csv.reader(f) 
    for line in reader:
        print(line)
```

**Multiple Reihen**

Mit Hilfe von `writerows` können auch mehre Zeilen auf einmal geschrieben werden. `writerows` bekommt ein Iterable of Iterables, zBsp eine Liste von Listen 

```python
with open("out_multi.csv", mode="w", newline='') as f: 
    writer = csv.writer(f)
    writer.writerows([["Eine", 1, "A", "Liste"], ["Zweite Reihe"]])
```

```python
open("out_multi.csv").read()
```

```python
with open("out_multi.csv") as f:
    reader = csv.reader(f) 
    for line in reader:
        print(line)
```

**Dict-Writer**

So wie es den Dictionary Reader gibt, gibt es auch einen Dictionary Writer.

Beim Erstellen des Writers müssen die Spaltennamen mit übergeben werden


```python
with open("out_dict.csv", mode="w", newline='') as f: 
    names = ["ID", "Punkte", "Kommentar"]
    writer = csv.DictWriter(f, names)
```

Der Header muss explizit geschrieben werden

```python
with open("out_dict.csv", mode="w", newline='') as f: 
    names = ["ID", "Punkte", "Kommentar"]
    writer = csv.DictWriter(f, names)
    writer.writeheader()
```

Zum Schreiben wird wieder die Methode `writerow` verwendet. Dieser muss jetzt jedoch ein Dictionary übergeben werden, mit gültigen Keys.

```python
with open("out_dict.csv", mode="w", newline='') as f: 
    names = ["ID", "Punkte", "Kommentar"]
    writer = csv.DictWriter(f, names)
    writer.writeheader()
    writer.writerow({"ID":4, "Punkte": 9, "Kommentar": "check"})
```

```python
open("out_dict.csv").read()
```

```python
with open("out_dict.csv") as f:
    reader = csv.reader(f) 
    for line in reader:
        print(line)
```

### Beispiel

```python
# Erstellen Sie eine mit ';' getrennte CSV-Datei mit den Spalten gerade;ungerade
# Befüllen Sie die CSV mit den Werte 1-50 entsprechend (bei einer geraden zahl ein Eintrag in der Spalte
# gerade, ansonsten ein Eintrag in der Spalte ungerade)
```

```python
with open("numbers.csv", mode="w", newline='') as f: 
    names = ["gerade", "ungerade"]
    writer = csv.DictWriter(f, names, delimiter=";")
    writer.writeheader()
    for i in range(1,51):
        if(i%2==0):
            writer.writerow({"gerade":i})
        else:
            writer.writerow({"ungerade":i})
```

```python
with open("numbers.csv") as f:
    reader = csv.reader(f) 
    for line in reader:
        print(line)
```

```python
# Erstellen Sie eine CSV-Datei mit den Spalten Ortsname, Kennzeichen, Bundesland
# Befüllen Sie Datei mit 4 Einträgen
```

```python
with open("orte.csv", mode="w", newline='', encoding='utf8') as f: 
    names = ["ortsname","kennzeichen","bundesland"]
    writer = csv.DictWriter(f, names)
    writer.writeheader()
    writer.writerow({"ortsname":"Wien","kennzeichen":"W","bundesland":"Wien"})
    writer.writerow({"ortsname":"St. Pölten","kennzeichen":"P","bundesland":"Niederösterreich"})
    writer.writerow({"ortsname":"Baden","kennzeichen":"BN","bundesland":"Niederösterreich"})
```

```python
with open("orte.csv", encoding='utf8') as f:
    reader = csv.reader(f) 
    for line in reader:
        print(line)
```

```python
# Erstellen Sie die gleiche CSV-Datei wie zuvor nur lassen Sie bei jeder Zeile einen Wert aus. 
# Wie werden leere Einträge behandelt? 
```

```python
with open("orte.csv", mode="w", newline='', encoding='utf8') as f: 
    names = ["ortsname","kennzeichen","bundesland"]
    writer = csv.DictWriter(f, names)
    writer.writeheader()
    writer.writerow({"kennzeichen":"W","bundesland":"Wien"})
    writer.writerow({"ortsname":"St. Pölten","bundesland":"Niederösterreich"})
    writer.writerow({"ortsname":"Baden","kennzeichen":"BN"})
```

```python
with open("orte.csv", encoding='utf8') as f:
    reader = csv.reader(f) 
    for line in reader:
        print(line)
```

## Seperator im Wert
Was passiert wenn ein Trennzeichen in einem Eintrag vorhanden ist?

zBsp ein **comma** seperiertes Feld hat den Eintrag: `ein, Beistrich`

Es gibt drei Lösungen für dieses Problem: 
 - Anderer Separator
 - Quotchar
 - Escape Character

**Delimiter**

```python
with open("quote.csv", mode="w", newline='') as f: 
    writer = csv.writer(f, delimiter = ';')
    writer.writerow(["1,3","2"])
```

```python
open("quote.csv").read()
```

**Quotchar**

```python
'1,3',2
```

```python
with open("quote.csv", mode="w", newline='') as f: 
    writer = csv.writer(f, quotechar="'")
    writer.writerow(["1,3","2"])
```

```python
open("quote.csv").read()
```

```python
with open("quote.csv", mode="w", newline='') as f: 
    writer = csv.writer(f, quotechar='"')
    writer.writerow(["1,3","2"])
```

```python
open("quote.csv").read()
```

**Escape Character** wird verwendet wenn keine Quotes verwendet werden

```python
with open("quote.csv", mode="w", newline='') as f: 
    writer = csv.writer(f, escapechar="!", quoting=csv.QUOTE_NONE)
    writer.writerow(",")
```

```python
open("quote.csv").read()
```

```python
with open("quote.csv", mode="w", newline='') as f: 
    writer = csv.writer(f, escapechar="!", quoting=csv.QUOTE_NONE)
    writer.writerow(["1,3","2"])
```

```python
open("quote.csv").read()
```

Lesen vom Weren mit **Escape Character**, wenn keine Quotes verwendet werden

```python
with open("quote.csv") as f:
    reader = csv.reader(f, escapechar="!")
    for line in reader:
        print(line)
```

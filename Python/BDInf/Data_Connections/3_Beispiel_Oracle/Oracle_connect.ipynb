# Oracle Database Connection

To connect to Oracle database you need to install Python cx_oracle 8 package.


## c_Oracle 8 Installation

To use cx_Oracle 8 with Python and Oracle Database you need:
- Oracle Client libraries. These can be downloaded from the free [Oracle Instant Client](https://www.oracle.com/database/technologies/instant-client.html)
- An Oracle Database, either local or remote

The cx_Oracle module loads Oracle Client libraries which communicate to an existing database.


## Run script

Set username and get password

```python
import getpass
username = 'flughafenbase'
userpwd = getpass.getpass("enter password")
```

##### initialize Oracle client library 

```python
import cx_Oracle

lib_dir=r".\instantclient_19_12_x64"

cx_Oracle.init_oracle_client(lib_dir=lib_dir)
```

##### Open connection to database

```python
dsn_tns = cx_Oracle.makedsn('localhost', '11521', service_name='XE') 
conn = cx_Oracle.connect(user=username, password=userpwd, dsn=dsn_tns) 
```

##### Execute query

```python
cursor = conn.cursor()
cursor.execute("""
        SELECT personid, vorname, nachname
        FROM person
        WHERE personid between :did AND :eid
        ORDER BY nachname""",
        did = 1,
        eid = 10)
```

##### Print result

```python
for pid, fname, lname in cursor:
    print("Personen:", fname, lname, "(", pid, ")")
```

### Pandas

```python
import pandas as pd

did = "1"
eid = "10"

query = """
        SELECT personid, vorname, nachname
        FROM person
        WHERE personid between """ + did + """ AND """ + eid + """
        ORDER BY nachname"""

df_ora = pd.read_sql(query, con=conn)
```

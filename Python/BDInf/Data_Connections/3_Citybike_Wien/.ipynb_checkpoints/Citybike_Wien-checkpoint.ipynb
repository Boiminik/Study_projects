# Citybike Wien Stationsabfrage

## Citybike Wien

https://citybikewien.at/de/

<p>Citybike Wien bietet an mehr als 120 Stationen in Wien Fahräder zum entleihen an. Sie können Ihr Rad an einer freien Bikebox einer beliebigen Station zurückgeben. Rund um die Uhr, das ganze Jahr lang!</p>

Über die URL https://dynamisch.citybikewien.at/citybike_xml.php?json können die aktuellen Stationsdaten als JSON Objekt abgefragt werden.

Nutzungsbedingungen: https://dynamisch.citybikewien.at/Nutzungsbedingungen_XML.pdf


## Auslesen der Daten über eine URL

* Um die aktuellen Stationsdaten über die URL abzufragen, wird die Bibliothek urllib importiert
* Zum verarbeiten des JSON Objekts wird die Bibliothek json importiert

```python
# import urllib library
from urllib.request import urlopen
  
# import json
import json
# store the URL in url as 
# parameter for urlopen
url = "https://dynamisch.citybikewien.at/citybike_xml.php?json"
  
# store the response of URL
response = urlopen(url)
  
# storing the JSON response 
# from url in data
bike_stations = json.loads(response.read())
```

### Ausgabe der Daten

```python
# print the json response
print(bike_stations)
```

### Formatierte Ausgabe der Daten

```python
# print the json response
print(json.dumps(bike_stations, indent=2))
```

## Darstellung der Stationen auf einer Karte

Um die Daten auf einer Karte darstellen zu können, muss gegebenenfalls das Modul folium installiert werden.

Folium Quickstart: https://python-visualization.github.io/folium/quickstart.html

```python
# Making a map using the folium module
try:
    import folium
except ImportError:
    !pip install folium
    import folium
```

### Marker der Karte hinzufügen

```python
# Center map on location Vienna
bike_map = folium.Map(location=[48.210033, 16.363449], zoom_start=12)


# Adding markers to the map
for bike_station in bike_stations:
    
    ## set location of station
    loc=bike_station['latitude'], bike_station['longitude']
    pop=bike_station['name'], bike_station['free_boxes'], bike_station['free_bikes']
    
    # set marker for station
    marker = folium.Marker(location=loc, popup=pop)
    
    # add marker to map
    marker.add_to(bike_map)


# Display map
bike_map
```

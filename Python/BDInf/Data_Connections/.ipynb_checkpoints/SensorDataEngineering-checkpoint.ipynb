Requires the following packages:
* pip install multiprocessing
* pip install psutil

```python
# Load the required packages
import psutil
import multiprocessing as mp

# Check the number of cores and memory usage
num_cores = mp.cpu_count()
print("This computer (kernel) has",num_cores,"cores")
print("Information regarding the memory usage:",psutil.virtual_memory())
```

```python
# some stats given by the psutil package
print("CPU usage:",psutil.cpu_times_percent())
print("Average load:",psutil.getloadavg())
print("Memory usage:",psutil.virtual_memory())
print("Battery state:",psutil.sensors_battery())
print("Network activities:",psutil.net_io_counters())
```

```python
import re
from time import sleep

for count in range(1,10):
    cpu_usage = psutil.cpu_percent()
    mem_used = str(psutil.virtual_memory())
    mem_used = re.search(" percent=([\d\.]+), ", mem_used).group(1)
    print("Measurement",count,"CPU:",cpu_usage,", Memory:",mem_used)
    sleep(5)
```

**Connect to InfluxDB and create DB/measurement using the influxdb package**

Requires the following Python packages:
* pip install influxdb

```python
from influxdb import InfluxDBClient

# connect to InfluxDB using default values
client = InfluxDBClient("localhost", 8086)
```

```python
# create a database called "MyTestDatabase"
client.create_database("MyTestDatabase")

# create a retention policy for this database: name of policy, duraction, replication, database, default, shard_duration
client.create_retention_policy('default_policy', '0s', 1, database="MyTestDatabase", default=True, shard_duration='7d')

# find documentation using:
# ?client.create_retention_policy
```

```python
# create dummy measurement (brushEvents) and dummy data
json_body = [
    {
        "measurement": "brushEvents",
        "tags": {
            "user": "Carol",
            "brushId": "6c89f539-71c6-490d-a28d-6c5d84c0ee2f"
        },
        "time": "2018-03-28T8:01:00Z",
        "fields": {
            "duration": 127
        }
    },
    {
        "measurement": "brushEvents",
        "tags": {
            "user": "Carol",
            "brushId": "6c89f539-71c6-490d-a28d-6c5d84c0ee2f"
        },
        "time": "2018-03-29T8:04:00Z",
        "fields": {
            "duration": 132
        }
    },
    {
        "measurement": "brushEvents",
        "tags": {
            "user": "Carol",
            "brushId": "6c89f539-71c6-490d-a28d-6c5d84c0ee2f"
        },
        "time": "2018-03-30T8:02:00Z",
        "fields": {
            "duration": 129
        }
    }
]

# write JSON data to database (and create measurements automatically!)
client.write_points(json_body, database = "MyTestDatabase")
```

```python
# close database connection again
client.close()
```

Store sensor data into InfluxDB 
-------------------------------

The following is shown in the upcoming cells:
* Creating a JSON snippet for one sensor data (via Python library)
* Store JSON snippet into an InfluxDB Measurement

**Task assignment: Run a loop and read in the CPU load and the available memory every 10 seconds; create a JSON snippet that is stored into the InfluxDB (database service has to be started beforehand)!**

```python
# libraries for retrieving OS metrics, datetime operations, JSONizing etc.
import re
import time
import datetime
import psutil
import json

# import library for InfluxDB connection
from influxdb import InfluxDBClient

# connect to InfluxDB using default values
client = InfluxDBClient("localhost", 8086)

# create a database called "OSMetricDB"
client.create_database("OSMetricDB")
```

```python
print("Starting to record OS metrics into the InfluxDB.")

# run a while loop until the stop button is pressed
try:
    while(True):
        cpu_usage = psutil.cpu_percent()
        mem_used = str(psutil.virtual_memory())
        mem_used = re.search(" percent=([\d\.]+), ", mem_used).group(1)
        print("Measurement - CPU:",cpu_usage,", Memory:",mem_used)
        # create a dictionary for this mesaurement
        entry = {
            "measurement": "osmetrics",
            "tags": {
                "host": "mycomputer",
                "user": "martin"
            },
            "time": str(datetime.datetime.now()),
            "fields": {
                "cpu": float(cpu_usage),
                "mem": float(mem_used)
            }
        }
        # write data point into the InfluxDB (we have to send a list of dicts!)
        client.write_points([entry], time_precision="s", database = "OSMetricDB")
        # wait 5 seconds
        time.sleep(5)

except KeyboardInterrupt:
    pass

print("Stopped recording of OS metrics!")
```

```python
# close DB connection again!
client.close()
```

Buffering and persisting sensor data
------------------------------------

The following is shown in the upcoming cells:
* Buffering a certain amount of sensor data in a dictionary
* Storing the JSONized dictionary into InfluxDB

```python
# some libraries we need for retrieving OS metrics, JSONizing etc.
import re
import time
import datetime
import psutil

# constant for storing a certain number of measurements
BUFFER_SIZE = 30

# import library for InfluxDB connection
from influxdb import InfluxDBClient

# connect to InfluxDB using default values
client = InfluxDBClient("localhost", 8086)

# create a database called "OSMetricDB"
client.create_database("OSMetricDB")
```

```python
print("Starting to record OS metrics into the InfluxDB.")

# run a while loop until the stop button is pressed
try:
    # list that should not contain more than BUFFER_SIZE entries
    buffer = []
    while(True):
        cpu_usage = psutil.cpu_percent()
        mem_used = str(psutil.virtual_memory())
        mem_used = re.search(" percent=([\d\.]+), ", mem_used).group(1)
        print("Measurement - CPU:",cpu_usage,", Memory:",mem_used)
        # create a dictionary for this mesaurement
        buffer.append({
            "measurement": "osmetrics",
            "tags": {
                "host": "mycomputer",
                "user": "martin"
            },
            "time": str(datetime.datetime.now()),
            "fields": {
                "cpu": float(cpu_usage),
                "mem": float(mem_used)
            }
        })
        # check if buffer is full; write data to InfluxDB if required
        if (len(buffer) >= BUFFER_SIZE):
            # write data point into the InfluxDB (we have to send a list of dicts!)
            client.write_points(buffer, time_precision="s", database = "OSMetricDB")
            # clear buffer
            buffer = []
            print("Buffered measurements were written to InfluxDB; buffer cleared...")
        # wait 1 second
        time.sleep(1)

except KeyboardInterrupt:
    pass

print("Stopped recording of OS metrics!")
```

```python
# close DB connection again!
client.close()
```

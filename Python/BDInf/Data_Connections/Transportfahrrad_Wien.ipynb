# Grätzlrad - Transportfahrrad-Sharing Standorte Wien Abfrage

## Grätlrad Wien

https://www.graetzlrad.wien/

<p>Die Transportfahrräder können Wienerinnen und Wiener kostenfrei ausleihen, um mit Einkäufe oder Kinder durch Wien zu radeln. Die Standorte sind über die ganze Stadt verteilt.</p>

Über die URL https://data.wien.gv.at/daten/geo?service=WFS&request=GetFeature&version=1.1.0&typeName=ogdwien:RADGRAETZELOGD&srsName=EPSG:4326&outputFormat=json können die aktuellen Stationsdaten als JSON Objekt abgefragt werden.

Quelle: https://www.data.gv.at/katalog/dataset/stadt-wien_graetzlrad--transportfahrradsharing-standorte-wien/resource/defed528-9a15-4268-8ad6-04880c84246e

Information: https://www.data.gv.at/katalog/dataset/stadt-wien_graetzlrad--transportfahrradsharing-standorte-wien

## Auslesen der Daten über eine URL

* Um die aktuellen Standortdaten über die URL abzufragen, wird die Bibliothek urllib importiert
* Zum verarbeiten des JSON Objekts wird die Bibliothek json importiert

```python
# import urllib library
from urllib.request import urlopen
  
# import json
import json
# store the URL in url as 
# parameter for urlopen
url = "https://data.wien.gv.at/daten/geo?service=WFS&request=GetFeature&version=1.1.0&typeName=ogdwien:RADGRAETZELOGD&srsName=EPSG:4326&outputFormat=json"
  
# store the response of URL
response = urlopen(url)
  
# storing the JSON response 
# from url in data
bike_stations = json.loads(response.read())
```

### Ausgabe der Daten

```python
# print the json response
print(bike_stations)
```

### Formatierte Ausgabe der Daten

```python
# print the json response
print(json.dumps(bike_stations, indent=2))
```

```python
folium.Map(location = [48.210033, 16.363449], zoom_start = 12)
```

### Marker einer Karte hinzufügen

#### Karte erstellen ####

```python
# Center map on location Vienna
bike_map = folium.Map(location = [48.210033, 16.363449], zoom_start = 12)
```

#### Daten auslesen, Marker erstellen und Karte hinzufügen ####

```python
# Adding markers to the map
tooltip = "Click me!"

# get data
for bike_station in bike_stations["features"]:
    
    ## set location of station
    loc = bike_station['properties']['HOCHWERT_WGS84'], bike_station['properties']['RECHTSWERT_WGS84']
    
    ## set popup information
    pop = '<b>Name:</b> '+ bike_station['properties']['BEZEICHNUNG'] + '<br>'\
        + '<b>URL:</b> <a href="' + bike_station['properties']['WEBLINK1'] + '">' + bike_station['properties']['WEBLINK1'] + '</a><br>' \
        + '<b>Type:</b> '+ bike_station['properties']['ICON'] + '<br>' \
        + '<b>Status:</b> '+ bike_station['properties']['STATUS']
    
    ## set marker for station
    marker = folium.Marker(location = loc, popup = pop, tooltip = tooltip)
    
    ## add marker to map
    marker.add_to(bike_map)

    ## Display data
    print(loc, pop)
```

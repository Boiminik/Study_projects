# Setup PostgreSQL

Setup the postgres Docker Container:

```python
!docker run -d --rm --name postgresdb -e POSTGRES_USER=user -e POSTGRES_PASSWORD=password -p 5432:5432 -v pgdata:/var/lib/postgresql/data postgres
```

Stop the Container (this will terminate the process):

```python
!docker stop postgresdb
```

Pause/Unpause the container (so you can use it later):

```python
!docker pause postgresdb
```

```python
!docker unpause postgresdb
```

Start the container:

```python
!docker start postgresdb
```

## Use the Docker Container

In the Docker Container Bash enter the following code to enter the Interface:

```python
!docker exec -it postgresdb bash
```

### Umwandeln in Exceldatei

Während read_* Funktionen verwendet werden, um Daten in Pandas zu lesen, werden die to_* Methoden verwendet, um Daten zu speichern. 

Die to_excel() Methode speichert die Daten als Excel-Datei. Im Beispiel wird mit dem Attribut sheet_name auf "Passagiere" anstelle des Standardbezeichnung Sheet1 gesetzt. Durch das Setzen index=False des Zeilenindex werden Beschriftungen nicht in der Kalkulationstabelle gespeichert.

titanic.to_excel("data/titanic.xlsx", sheet_name="passengers", index=False)

```python
#### Lesen der neu erstellten Exceldatei
```

titanic_excel = pd.read_excel("data/titanic.xlsx", sheet_name="passengers")

### Zusammenfassung des Dataframes mit der Methode info()

titanic_excel.info()

* Es ist ein DataFrame.
* Es gibt 891 Einträge, d.h. 891 Zeilen.
* Jede Zeile hat eine Zeilenbeschriftung (auch bekannt als index) mit Werten von 0 bis 890.
* Die Tabelle hat 12 Spalten. Die meisten Spalten haben einen Wert für jede der Zeilen (alle 891 Werte sind non-null). Einige Spalten enthalten fehlende Werte und weniger als 891 non-nullWerte.
* Die Spalten Name, Sex, Cabin und Embarked bestehen aus Textdaten (Strings, auch bekannt als object). Die anderen Spalten sind numerische Daten, von denen einige ganze Zahlen (aka integer) und andere reelle Zahlen (aka float) sind.
* Die Art der Daten (Zeichen, Ganzzahlen,…) in den verschiedenen Spalten werden durch Auflisten der dtypes.
* Die ungefähre Menge an RAM, die zum Halten des DataFrame verwendet wird, wird ebenfalls angegeben.

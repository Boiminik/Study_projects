#### Work with SQL and JSON on a Postgres-Database
Follow these steps:

	1. use "docker-compose" of the file "Postgres-Docker.zip" to start a Postgres Database Server
	2. start DataGrip and set a connection to the Postgres Server (see credentials in the docker-compose file)
	3. copy music-store.sql from host to container
	4. execute music-store.sql and create music-store tables
	5. check tables in Postgres server in DataGrip
	6. copy library.json from host to container
	7. create table jtracks and copy data from library.json
	8. check tables in Postgres server in DataGrip
	9. remove copied files from container
	10. open a console window in the Postgres Server in DataGrip
	11. to get a better overview about the tables (we will actually need only a few of them) check out the "music-store-DB-diagram"
	12. copy all statements of the file "JSON-Demo.sql" to the DataGrip console and play with the given select statements (which allow to address the JSON parts of the data)
	13. watch the changes done from Python-Code in DataGrip

#### Create music-store schema

```python
# copy music-store.sql from host to container
!docker cp music-store.sql postgres-db-1:/music-store.sql
```

```python
# Create music-store tables
!docker exec -i postgres-db-1 bash -c "psql -U admin -d postgres < music-store.sql"
```

#### Create jtracks table

```python
# copy library.json from host to container
!docker cp library.json postgres-db-1:/library.json
```

```python
# Create table jtraks in postgres db
!docker exec -i postgres-db-1 bash -c "psql -U admin -d postgres -c 'CREATE TABLE IF NOT EXISTS jtracks(id SERIAL, body JSONB);'"
```

```python
# copy json data to table
!docker exec -i postgres-db-1 bash -c "psql -U admin -d postgres -c \"\copy jtracks(body)FROM '/library.json' WITH CSV QUOTE E'\x01' DELIMITER E'\x02';\""
```

```python
# delete copied file from container
!docker exec -i postgres-db-1 bash -c "rm /music-store.sql /library.json"
```

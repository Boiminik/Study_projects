```python
# Step 1: Start MySQL server using Docker Compose

import os

# Navigate to the directory containing the docker-compose.yml file
docker_compose_dir = r'C:\Users\Startklar\Documents\Study\BDInf\JupyterNotebook\MySQL\docker_compose'
os.chdir(docker_compose_dir)

# Execute docker-compose up command to start MySQL server
os.system('docker-compose up -d')

# Wait for a few seconds for the server to start
import time
time.sleep(10)

# Step 2: Connect to the MySQL server using mysql-connector

import mysql.connector

# Define MySQL connection parameters
mysql_host = 'localhost'
mysql_port = 3306
mysql_user = 'root'
mysql_password = 'password'
mysql_database = 'airportbase'

# Connect to MySQL server
try:
    connection = mysql.connector.connect(
        host=mysql_host,
        port=mysql_port,
        user=mysql_user,
        password=mysql_password,
        database=mysql_database
    )
    print("Connected to MySQL server successfully!")
except mysql.connector.Error as err:
    print(f"Error: {err}")

# Step 3: Execute SQL queries and monitor changes

# You can execute SQL queries and monitor changes directly within Jupyter Notebook.
# For example:

# Execute a simple query to fetch some data
try:
    cursor = connection.cursor()
    cursor.execute("SELECT * FROM DEPARTMENT;")
    result = cursor.fetchall()
    
    # Display the fetched data
    for row in result:
        print(row)
        
except mysql.connector.Error as err:
    print(f"Error: {err}")

# Close the cursor and connection
cursor.close()
connection.close()
```

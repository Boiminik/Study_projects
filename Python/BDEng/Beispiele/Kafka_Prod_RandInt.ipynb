```python
# load important libraries
from time import sleep
from json import dumps
from kafka import KafkaProducer
import random
import datetime;

# create producer
producer = KafkaProducer(bootstrap_servers=['kafka:9092'],
                         value_serializer=lambda x: dumps(x).encode('utf-8'))

```

```python
# now write the numbers 1 to 2000 to an existing topic
# (with a delay of random seconds)
for e in range(2000):   
    data = {'number' : e, 'rand' : random.randint(0, 6), 'time' : datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")}
    print('streaming: ', data)
    producer.send('randTopic', value=data)
    sleep(random.randint(1, 5))
```

```python
# delete Topic
from kafka.admin import KafkaAdminClient

admin_client = KafkaAdminClient(
    bootstrap_servers="kafka:9092", 
    client_id='test')

admin_client.delete_topics(['randTopic'])
```

# Spark Output Sinks

This Notebook uses "Kafka_Prod_RandInt.ipynb" as Kafka Producer for the data uses here

### Spark initialization

```python
from pyspark.sql import SparkSession

# Spark session & context
spark = (SparkSession
         .builder
         .appName('RandInt')
         .config("spark.jars.packages", "org.apache.spark:spark-sql-kafka-0-10_2.12:3.3.0")
         .getOrCreate())
sc = spark.sparkContext
```

### initialize readStream

```python
# Create stream dataframe setting kafka server, topic and offset option
df = (spark
  .readStream
  .format("kafka")
  .option("kafka.bootstrap.servers", "kafka:9092") # kafka server
  .option("subscribe", "randTopic") # topic
  .option("startingOffsets", "earliest") # start from beginning 
  .option("trigger","1 seconds")
  .option("maxOffsetsPerTrigger",10) #Rate limit on maximum number of offsets processed per trigger interval
  .load())
```

### prepare data (create schema and do initial data conversions)

```python
# Convert binary to string key and value
from pyspark.sql.functions import from_json
from pyspark.sql.types import StructType, StructField, BooleanType, LongType, IntegerType, StringType

# Event data schema
data_schema = StructType(
    [StructField("number",IntegerType(),True),
     StructField("rand",IntegerType(),True),
     StructField("time",StringType(),True),
    ])

df1 = (df
    .withColumn("key", df["key"].cast(StringType()))
    .withColumn("value", df["value"].cast(StringType()))
    .withColumn("value", from_json("value", data_schema))
    )

display(df1)
```

```python
from pyspark.sql.functions import col, from_unixtime, to_date, to_timestamp

# Transform into tabular 
# Convert unix timestamp to timestamp
# Create partition column (change_timestamp_date)
df1_formatted = (df1.select(
    col("key").alias("event_key"),
    col("topic").alias("event_topic"),
    col("timestamp").alias("event_timestamp"),
    col("value.number").alias("number"),
    col("value.rand").alias("rand"),
    col("value.time").alias("time")
))
display(df1_formatted)
```

## Spark output sinks

```python
# Output Sink JSON
df1_OutputStream_json = (df1_formatted
   .writeStream
   .queryName("streamOutjson")                                     # Name the query
   .trigger(processingTime="15 seconds")
   .format("json")       # supports these formats : csv, json, orc, parquet
   .option("path", "./file_sink_json1")
   .option("header", True)
   .option("checkpointLocation", "checkpoint_json1")
   .outputMode("append") # Filesink only support Append mode.
   .start()
   #.awaitTermination()
 )
```

```python
# Output Sink CSV
df1_OutputStream_csv = (df1_formatted
   .writeStream
   .outputMode("append") # Filesink only support Append mode.
   .format("csv")       # supports these formats : csv, json, orc, parquet
   .option("path", "./file_sink_csv1")
   .option("header", True)
   .option("checkpointLocation", "checkpoint_csv1")
   .start()
   #.awaitTermination()
 )
```

```python
# Output Sink - foreach with class
# does this really work??

class ForeachWriter:
    
    def open(self, partition_id, epoch_id):
        # Open connection. This method is optional in Python.
        self.f = open('csvfile.csv','a')
        print("Opened %d, %d" % (partition_id, epoch_id))
        return True

    def process(self, row):
        # Write row to connection. This method is not optional in Python.
        self.f.write(row+'\n') #Give your csv text here.
        print(row)

    def close(self, error):
        # Close the connection. This method is optional in Python.
        print("Closed with error: %s" % str(error))
        self.f.close()        

df1_OutputStream_csv_foreach = (df1_formatted
   .writeStream
   .foreach(ForeachWriter())
   .start()
   #.awaitTermination()
 )
```

```python
# Output Sink - foreach with function
# same here: does this work?
def print_row(row):
    print(row)

writer = (df1_formatted
          .writeStream
          .foreach(print_row)
          .start())
```

# work with data

```python
# transform to windowed data
from pyspark.sql.functions import *

tumblingWindows = (df1_formatted
                   .withWatermark("event_timestamp", "10 minutes")
                   .groupBy("number", window("event_timestamp", "5 minutes"))
                   .count()
                  )

display(tumblingWindows)
print(tumblingWindows.isStreaming)
```

```python
# register data as writeStream to memory
df1_OutputStream_memory = (tumblingWindows
   .writeStream
   .queryName("streamOutMemory")                                     # Name the query
   .outputMode("update") # Filesink only support Append mode.
   .format("memory")       # supports these formats : csv, json, orc, parquet
   .start()
   #.awaitTermination()
 )
```

```python
# make selects on the memory stream
from pyspark.sql.functions import *

tumblingWindows1 = spark.sql(
    """select * from streamOutMemory;
    """)
tumblingWindows1.show(truncate = False)
```

```python
from kafka import KafkaConsumer
import json

consumer = KafkaConsumer(
 bootstrap_servers='kafka:9092',
 value_deserializer = lambda v: json.loads(v.decode('ascii')),
 auto_offset_reset='earliest'
)
```

```python
consumer.subscribe(topics='testTopic')
for m in consumer:
  print ("%d:%d: v=%s" % (m.partition,
                          m.offset,
                          m.value['number']))
```

```python
consumer.subscribe(topics='wiki-changes')
for m in consumer:
  print ("%d:%d: user=%s bot=%s" % (m.partition,
                          m.offset,
                          m.value['user'],
                          m.value['bot']))
```

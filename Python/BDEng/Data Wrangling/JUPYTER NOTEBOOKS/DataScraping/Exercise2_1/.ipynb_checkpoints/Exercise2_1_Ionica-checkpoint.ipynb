```python
# Import libraries
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time
import pandas as pd
import json
```

```python
# get the webpage content
driver = webdriver.Chrome()
url = "https://www.kaggle.com/code/parulpandey/geek-girls-rising-myth-or-reality/comments"
driver.get(url)
```

```python
comments_data = []
comments = driver.find_elements(By.CLASS_NAME, "sc-DnZRP") 
```

```python
for item in comments:
    print(item)
```

```python
for comment in comments:
    # Extract the author's name
    author_name = comment.find_element(By.CSS_SELECTOR, "h3.sc-fHejqy").text 
    # Extract timestamp
    temp = comment.find_element(By.XPATH,'//*[@id="670369"]/div/div[1]/div[1]/div/div/div[2]/p/span')
    timestamp = temp.get_attribute('title')
    # Extract the comment text
    div_element = comment.find_element(By.CSS_SELECTOR, ".sc-kGLCbq.fAhUHg")
    comment_text = div_element.find_element(By.TAG_NAME, 'p').text
   
    # Prepare dictionary
    comment_info = {
        "author": author_name,
        "timestamp": timestamp,
        "comment_text": comment_text
    }
    # Append the comment 
    comments_data.append(comment_info)
    
```

```python
print(comments_data)
```

```python
# Quit the driver after the scraping is complete
driver.quit()
```

```python
# Write the comments data to a JSON file
with open('kaggle_comments.json', 'w', encoding='utf-8') as f:
    json.dump(comments_data, f, indent=4)
```

```python
import requests
import re
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = requests.get('https://stackoverflow.com/questions/2081586/web-scraping-with-python')
```

```python
soup = BeautifulSoup(url.text, 'html.parser')
```

```python
# collect needed data
used_tags = ['div', 'span']
action_class = 'user-action-time fl-grow1' #tag: div
rating_class = 'js-vote-count flex--item d-flex fd-column ai-center fc-theme-body-font fw-bold fs-subheading py4' #tag: div
username_itemprop = 'name' #tag: span
timestamp_class = 'relativetime' #tag: span
textBody_class = 's-prose js-post-body' #tag: div
```

```python
# set find all here
action_tags = soup.find_all(used_tags[0], {'class': action_class})
textBody_tags = soup.find_all(used_tags[0], {'class': textBody_class})
username_tags = soup.find_all(used_tags[1], {'itemprop': username_itemprop})
timestamp_tags = soup.find_all(used_tags[1], {'class': timestamp_class})
rating_tags = soup.find_all(used_tags[0], {'class': rating_class})
```

```python
#get number of answer posts
post_count_text = soup.find('span', {'itemprop':'answerCount'}) 
pcount = int(post_count_text.text)
#add +1 for the question post
pcount +=1
```

```python
posts = []
```

```python
# Loop over the range of pcount
for instance_number in range(1, pcount):
    # Extracting action text without 'edited'
    action_text = None
    for tag in action_tags:
        if 'edited' not in tag.text.lower():
            action_text = tag.text.strip()
            action_tags.remove(tag)  # Remove this tag to avoid reusing it
            break

    # Extracting data for each instance
    if action_text:
        action_parts = action_text.split()
        if len(action_parts) >= 6:
            action_date = f"{action_parts[-2]} {action_parts[-1]} {action_parts[3]}"
            action_time = action_parts[5]
        else:
            action_date = action_time = None

        timestamp_text = timestamp_tags[instance_number - 1]['title'].strip() if instance_number <= len(timestamp_tags) else None
        timestamp_parts = timestamp_text.split(', ') if timestamp_text else ['', '']

        posts.append((
            action_parts[0],  # action
            action_parts[2].strip(','),  # day without comma
            action_parts[1],  # month
            action_parts[3],  # year
            action_time,  # time
            textBody_tags[instance_number - 1].text.strip() if instance_number <= len(textBody_tags) else None,
            username_tags[instance_number - 1].text.strip() if instance_number <= len(username_tags) else None,
            rating_tags[instance_number - 1].text.strip() if instance_number <= len(rating_tags) else None
        ))

# Now the 'posts' list contains all the extracted data without any additional strings
print(posts)
```

```python
posts_df = pd.DataFrame(posts, columns=['action', 'action_day', 'action_month', 'action_year', 'action_time', 'text_body', 'username', 'user_ratings'])
```

```python
posts_df.to_csv('StackOverflow_WebScraping-Posts.csv')
```

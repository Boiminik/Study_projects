```python
item_classes = ['Item-item-S33', 'Item-item-J04']
price_class = 'Text-bold-t5X'
```

```python
houses = []
```

```python
for p in plz:
    r = requests.get(f'https://www.immobilienscout24.at/regional/{p}/wohnung-kaufen')

    soup = BeautifulSoup(r.text, 'html.parser')
    count = soup.find('span', {'class': 'headline-count'})

    print(p, count.text)

    houses_s33 = soup.find_all('li', {'class': item_classes[0]})
    houses_j04 = soup.find_all('li', {'class': item_classes[1]})

    for house in houses_s33 + houses_j04:
        price = house.find('li', {'class': price_class})
        houses.append({'plz': p, 'price': price.text})
```

```python
import pandas as pd
```

```python
houses_df = pd.DataFrame(houses)
```

```python
houses_df.to_csv('immoscout.csv')
```

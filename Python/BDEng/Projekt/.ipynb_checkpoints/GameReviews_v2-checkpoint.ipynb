### Introduction
In this project, we aim to compare user reviews with journalist reviews to uncover any potential discrepancies or biases. We will collect a comprehensive dataset of reviews from OpenCritic, focusing initially on those authored by journalists from IGN. Through systematic analysis, we seek to highlight significant differences between professional and user-generated reviews, which may reveal biases or irregularities within professional critiques. Our ultimate goal is to ensure greater transparency and accountability in the review process, fostering a more trustworthy gaming review ecosystem.

First, we need to gather the OpenCritic game ID for each game we want the review score of, to do that, we iterate through our games array and send each string to the API to search for the game ID:

```python
# API key used for the API access
api_key = "64686b9a75msha576b74f8ff0b56p1a63ebjsnf20f69b4fac9"

# Function to get game IDs
def get_game_ids(game_names, api_key):
    all_game_ids = []
    filtered_game_names = []

    for game_name in game_names:
        encoded_game_name = urllib.parse.quote(game_name)
        url = f"https://opencritic-api.p.rapidapi.com/game/search?criteria={encoded_game_name}"

        headers = {
            "X-RapidAPI-Host": "opencritic-api.p.rapidapi.com",
            "X-RapidAPI-Key": api_key
        }

        response = requests.get(url, headers=headers)

        if response.status_code == 200:
            data = response.json()
            if data:
                # Check if the name of the top search result matches the game name exactly
                top_game = data[0]
                if top_game['name'].lower() == game_name.lower():
                    top_game_id = top_game['id']
                    all_game_ids.append((game_name, top_game_id))
                    filtered_game_names.append(game_name)
                    print(f"Game ID for '{game_name}' is {top_game_id}")
                else:
                    print(f"No exact match found for '{game_name}'")
            else:
                print(f"No results found for '{game_name}'")
        else:
            print(f"Error searching for '{game_name}': ", response.status_code)

    return filtered_game_names, all_game_ids

# Call the method
filtered_game_names, game_ids = get_game_ids(game_names, api_key)

# Save game IDs to JSON file
data = {"filtered_game_names": filtered_game_names, "game_ids": game_ids}
with open("data/game_ids.json", "w") as json_file:
    json.dump(data, json_file)

print("Game IDs saved to data/game_ids.json")
```

Now we are going to collect all the needed Data for each of the outlet IDs as well as the average score from every critic from every game. Since the API only collects 20 Reviews at a Time, we will need a loop to go through 20 sets, then skip the 20 already found and start a new search, until we have gathered each review for each game:

```python
# API key used for the API access
api_key = "64686b9a75msha576b74f8ff0b56p1a63ebjsnf20f69b4fac9"

# List of outlet IDs
outlets = {
    "IGN": 56,
    "PC Gamer": 162,
    "GamesRadar+": 91,
    "Metro GameCentral": 75,
    "Eurogamer": 114,
    "Easy Allies": 394,
    "Game Informer": 35,
    "Forbes": 290,
    "GameSpot": 32,
    "Game Rant": 60
}

# Function to load platforms data from platforms.json
def load_platforms_data(filename="platforms.json"):
    with open(filename, 'r') as f:
        platforms_data = json.load(f)
    platforms_map = {platform['id']: platform['name'] for platform in platforms_data}
    return platforms_map

# Function to get reviews for each game ID
def get_reviews_for_games(game_ids, api_key, sort_by="newest"):
    all_reviews = []

    for game_name, game_id in game_ids:
        game_reviews = []
        skip = 0

        while True:
            url = f"https://opencritic-api.p.rapidapi.com/reviews/game/{game_id}?skip={skip}&sort={sort_by}"

            headers = {
                "X-RapidAPI-Host": "opencritic-api.p.rapidapi.com",
                "X-RapidAPI-Key": api_key
            }

            response = requests.get(url, headers=headers)

            if response.status_code == 200:
                data = response.json()
                if not data:
                    break  # No more reviews to fetch
                game_reviews.extend(data)
                skip += 20
            else:
                print(f"Error for game ID '{game_id}' at skip={skip}: ", response.status_code)
                break

        all_reviews.extend([(game_name, review) for review in game_reviews])
        print(f"Collected {len(game_reviews)} reviews for game '{game_name}'")

    return all_reviews

# Function to process reviews and save to JSON
def process_and_save_reviews(reviews, outlets, platforms_map, path="data/opencritic_reviews.json"):
    result = {}

    # Dictionary to store platform scores and counts
    platform_scores = {}
    platform_counts = {}

    for game_name, review in reviews:
        outlet_id = review['Outlet']['id']
        outlet_name = review['Outlet']['name']
        score = review.get('score')
        created_at = review['createdAt'][:10]  # Extract the date part (YYYY-MM-DD)
        authors = review.get('Authors', [])
        author_name = authors[0]['name'] if authors else None

        # Extract platforms from the review
        platforms = review.get('Platforms', [])
        if platforms:
            for platform in platforms:
                platform_id = platform['id']
                if platform_id in platforms_map:
                    platform_name = platforms_map[platform_id]
                    if platform_name not in platform_scores:
                        platform_scores[platform_name] = []
                        platform_counts[platform_name] = 0
                    if isinstance(score, (int, float)):
                        platform_scores[platform_name].append(score)
                        platform_counts[platform_name] += 1

        if game_name not in result:
            result[game_name] = {"outlets": {}, "platforms": {}, "average_score": None}

        if outlet_id in outlets.values():
            # Check if the score is a valid number
            if isinstance(score, (int, float)) and score is not None:
                result[game_name]["outlets"][outlet_name] = {
                    "score": score,
                    "review_url": review['externalUrl'],
                    "created_at": created_at,
                    "author": author_name  # Include author's name if available
                }

    # Calculate average scores for each platform
    for platform_name, scores in platform_scores.items():
        if scores:
            avg_score = sum(scores) / len(scores)
            result[game_name]["platforms"][platform_name] = {
                "average_score": avg_score,
                "count": platform_counts[platform_name]
            }

    # Calculate average score for the game
    all_scores = [outlet_data["score"] for outlet_data in result[game_name]["outlets"].values() if isinstance(outlet_data["score"], (int, float))]
    if all_scores:
        result[game_name]["average_score"] = sum(all_scores) / len(all_scores)

    # Save to JSON
    with open(path, 'w') as json_file:
        json.dump(result, json_file, indent=4)
    print(f"Reviews data saved to {path}")

# Load game IDs from JSON file
with open("data/game_ids.json", "r") as json_file:
    data = json.load(json_file)
    game_ids = data["game_ids"]

# mocking data; for testing
# game_ids = [["Elden Ring", 12090]]

# Load platforms data
platforms_map = load_platforms_data("data/platforms.json")

# Get reviews for each game
all_reviews = get_reviews_for_games(game_ids, api_key)

# Process reviews and save to JSON
process_and_save_reviews(all_reviews, outlets, platforms_map)
```

## Web Scraping

```python
# Load game names from JSON file
with open("data/game_ids.json", "r") as json_file:
    df_names = json.load(json_file)
    filtered_game_names = df_names["filtered_game_names"]

# List of game names to scrape (parameter: filtered_game_names) mocking data
# filtered_game_names = [
    # "grand theft AUTO v",
    # "Elden Ring",
    # "Wii Sports",
    # "The Legend Of Zelda: Breath of The Wild",
    # "Cooking Mama",
    # "StardEW valley",
    # "Resident Evil 5",
    # "Unga Bunga doogaloo"
    # Add more game names here
# ]

# Base URL
base_url = "https://www.backloggd.com/games/"

# Initialize a list to store game names, scores, plays, and genres
data = []

# Loop through each game name in the list
for game_name in filtered_game_names:
    # Remove any colons from the game name
    game_name_clean = game_name.replace(':', '')
    
    # Create the URL by replacing spaces with hyphens and converting to lowercase
    url = f"{base_url}{game_name_clean.replace(' ', '-').lower()}/"
    
    # Send a GET request to the URL
    response = requests.get(url)
    
    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # Find the div with id 'score'
        score_div = soup.find('div', id='score')
        
        # Initialize variables for score, plays, and genres
        score_percentage = None
        plays_count = None
        genres = []
        
        # If the score_div is found, extract the score
        if score_div:
            score = score_div.find('h1').text.strip()
            try:
                # Convert score to float and divide by 5 to get the percentage
                score_percentage = float(score) / 5 * 100
            except ValueError:
                # Handle the case where the score is not a valid number
                print(f"Invalid score for game: {game_name}")
        
        # Find the plays counter
        plays_counter = soup.find('a', class_='plays-counter', href=True)
        
        # If the plays_counter is found, extract the plays count
        if plays_counter:
            plays_text = plays_counter.find_all('p')[-1].text.strip()
            plays_count = plays_text
        
        # Find all genre tags
        genre_tags = soup.find_all('p', class_='genre-tag')
        
        # Extract genres if they exist
        for genre_tag in genre_tags:
            genre = genre_tag.text.strip()
            genres.append(genre)
        
        # Only append if at least one of score, plays, or genres is not None
        if score_percentage is not None or plays_count is not None or genres:
            data.append({
                'Game Name': game_name, 
                'Score': score_percentage, 
                'Plays': plays_count, 
                'Genres': ', '.join(genres) if genres else None
            })
    else:
        print(f"Failed to retrieve data for game: {game_name}")

# Create a DataFrame from the data list
df = pd.DataFrame(data)

# Save the DataFrame to a Parquet file
df.to_parquet('data/user_scores.parquet', index=False)

print("Data saved to user_scores.parquet")
```

Now let's take a look at the actual file we just created, there may be some wrongly gathered data:

```python
# Read the Parquet file
df = pd.read_parquet('data/user_scores.parquet')

print(df.count())
print(df[df.isnull().any(axis=1)])
# Display the DataFrame
print(df.min())
print(df.head())
```

Looks like we got three datasets with NaN value, lets remove those. Also Looks like the score is not consistent with our previously gathered dataframes, let's round these so we can get integers. The plays column is also very inconsistent, lets remove the 'k' and multiply the number by 1000 to get consistent numbers:

```python
# Drop all NAs from the DataFrame
df.dropna(inplace=True)

# Round the scores to integers
df['Score'] = df['Score'].round().astype(int)

# Function to process the plays count
def process_plays(plays):
    if pd.isna(plays):
        return plays
    if 'K' in plays:
        return int(float(plays.replace('K', '')) * 1000)
    return int(plays)

# Apply the function to the Plays column
df['Plays'] = df['Plays'].apply(process_plays)

# Save the cleaned DataFrame to a new Parquet file
df.to_parquet('data/cleaned_user_scores.parquet', index=False)

print("Cleaned data saved to cleaned_user_scores.parquet")
```

### Kafka/Spark alternative

There seems to be a client side error that prevents us from using Kafka/Spark to edit the data, here is the code for this anyway, it should work on a suitable device:

```python
# Initialize SparkSession
spark = SparkSession.builder \
    .appName("ReadParquetAndStreamToKafka") \
    .getOrCreate()

# Read Parquet file into a DataFrame
df = spark.read.parquet("data/user_scores.parquet")

# Show the DataFrame schema and content
df.printSchema()
df.show(truncate=False)
```

```python
# Write data to Kafka topic
kafka_broker = "localhost:29092"
kafka_topic = "user_scores_topic"

df.withColumnRenamed("Game Name", "GameName") \
  .selectExpr("CAST(GameName AS STRING) AS key", "to_json(struct(*)) AS value") \
  .write \
  .format("kafka") \
  .option("kafka.bootstrap.servers", kafka_broker) \
  .option("topic", kafka_topic) \
  .save()
```

```python
# Initialize SparkSession
spark = SparkSession.builder \
    .appName("KafkaSparkStreaming") \
    .getOrCreate()

# Define schema for Kafka message
schema = StructType() \
    .add("Game Name", StringType()) \
    .add("Score", DoubleType()) \
    .add("Plays", StringType()) \
    .add("Genres", StringType())

# Read from Kafka
kafka_stream = spark.readStream \
    .format("kafka") \
    .option("kafka.bootstrap.servers", "localhost:29092") \
    .option("subscribe", "user_scores_topic") \
    .load()

# Convert value from binary to string
kafka_stream = kafka_stream.selectExpr("CAST(value AS STRING)")

# Parse JSON data
parsed_stream = kafka_stream.select(from_json(col("value"), schema).alias("data")).select("data.*")

# Drop NA values
parsed_stream = parsed_stream.dropna()

# Round the scores
parsed_stream = parsed_stream.withColumn("Score", expr("ROUND(Score)").cast(IntegerType()))

# Function to process plays count
def process_plays(plays):
    if 'K' in plays:
        return int(float(plays.replace('K', '')) * 1000)
    return int(plays)

# Apply the function to Plays column
processed_stream = parsed_stream.withColumn("Plays", expr("process_plays(Plays)").cast(IntegerType()))

# Write processed data to Parquet file
query = processed_stream.writeStream \
    .outputMode("append") \
    .format("parquet") \
    .option("checkpointLocation", "checkpoints/user_scores") \
    .start("data/cleaned_user_scores.parquet")

# Await termination
query.awaitTermination()
```

## Average Ratings by Platform

In this section, we analyze the average ratings given by journalists for each platform. This analysis helps us identify any significant differences in ratings across various gaming platforms.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import json

# Load the reviews data
with open('data/opencritic_reviews.json', 'r') as file:
    reviews_data = json.load(file)

# Prepare a dictionary to hold platform scores
platform_scores = {}

# Extract platform scores from the reviews data
for game, details in reviews_data.items():
    if 'platforms' in details:
        for platform, platform_info in details['platforms'].items():
            if 'average_score' in platform_info:
                if platform not in platform_scores:
                    platform_scores[platform] = []
                platform_scores[platform].append(platform_info['average_score'])

# Calculate the average score for each platform
average_platform_scores = {platform: sum(scores) / len(scores) for platform, scores in platform_scores.items()}

# Convert the results to a DataFrame
platform_scores_df = pd.DataFrame(list(average_platform_scores.items()), columns=['Platform', 'Average_Score'])

# Filter out platforms with no valid scores
platform_scores_df = platform_scores_df.dropna()

# Sort by average score
platform_scores_df = platform_scores_df.sort_values(by='Average_Score', ascending=False)

# Create the bar plot with friendlier colors
plt.figure(figsize=(12, 8))
plt.barh(platform_scores_df['Platform'], platform_scores_df['Average_Score'], color='cornflowerblue')
plt.xlabel('Average Score')
plt.title('Average Journalist Ratings by Platform')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

The bar chart shows the average journalist ratings for each platform. The platforms are sorted in descending order of their average ratings.

The results indicate that the average ratings for different platforms are quite similar, all being very high. Nintendo Switch and PlayStation 5 have the highest average ratings, followed by Xbox Series X/S and PC. This suggests that these platforms may offer higher quality games or that the games on these platforms are rated more favorably by journalists.

This visualization helps us recognize trends in journalist ratings for different platforms and understand how the ratings vary across the platforms.

### Average Ratings by Publisher with Number of Reviews
For the next part, we will combine the average ratings by publisher with the number of reviews to gain further insights. This will help us understand how the volume of reviews might impact the average ratings for each publisher.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the games data
games_df = pd.read_csv('data/games.csv')

# Convert the 'Number of Reviews' column to numeric, errors='coerce' will replace non-numeric values with NaN
games_df['Number of Reviews'] = pd.to_numeric(games_df['Number of Reviews'], errors='coerce')

# Drop rows with NaN values in 'Number of Reviews' or 'Rating' columns
games_df = games_df.dropna(subset=['Number of Reviews', 'Rating'])

# Convert 'Rating' and 'Number of Reviews' to numeric
games_df['Rating'] = pd.to_numeric(games_df['Rating'], errors='coerce')
games_df['Number of Reviews'] = pd.to_numeric(games_df['Number of Reviews'], errors='coerce')

# Drop rows with NaN values in 'Rating' or 'Number of Reviews' columns
games_df = games_df.dropna(subset=['Rating', 'Number of Reviews'])

# Calculate the cumulative rating and the number of reviews for each publisher
publisher_stats = games_df.groupby('Team').agg(
    Total_Rating=('Rating', 'sum'),
    Total_Reviews=('Number of Reviews', 'sum')
).reset_index()

# Clean the publisher names
publisher_stats['Team'] = publisher_stats['Team'].str.replace(r"[\'\[\]]", "", regex=True)

# Calculate the average rating for each publisher
publisher_stats['Average Rating'] = publisher_stats['Total_Rating'] / publisher_stats['Total_Reviews']

# Sort by the average rating
publisher_stats = publisher_stats.sort_values(by='Average Rating', ascending=False)

# Select the top 10 and bottom 10 publishers
top_10_publishers = publisher_stats.head(10)
bottom_10_publishers = publisher_stats.tail(10)

# Plot the top 10 publishers
plt.figure(figsize=(14, 10))
bars = plt.barh(top_10_publishers['Team'], top_10_publishers['Average Rating'], color='skyblue')
plt.xlabel('Average Rating')
plt.title('Top 10 Average Ratings by Publisher with Total Reviews')
plt.gca().invert_yaxis()

# Add text labels to each bar
for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.1, bar.get_y() + bar.get_height() / 2, f'{width:.1f}', ha='left', va='center')

plt.subplots_adjust(left=0.3, right=0.9, top=0.9, bottom=0.1)
plt.show()

# Plot the bottom 10 publishers
plt.figure(figsize=(14, 10))
bars = plt.barh(bottom_10_publishers['Team'], bottom_10_publishers['Average Rating'], color='salmon')
plt.xlabel('Average Rating')
plt.title('Bottom 10 Average Ratings by Publisher with Total Reviews')
plt.gca().invert_yaxis()

# Add text labels to each bar
for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.1, bar.get_y() + bar.get_height() / 2, f'{width:.1f}', ha='left', va='center')

plt.subplots_adjust(left=0.3, right=0.9, top=0.9, bottom=0.1)
plt.show()
```

The analysis of average user ratings by publisher provides valuable insights for our project. By identifying the top and bottom publishers, we can better understand the factors contributing to high and low user satisfaction.

- **Top Publishers**: Publishers like **Sony Computer Entertainment**, **FromSoftware**, and **ZA/UM** are consistently producing highly-rated games. Examining their review patterns and user feedback can reveal the factors leading to high ratings, such as innovative game mechanics, engaging storylines, or high-quality graphics.

- **Bottom Publishers**: Publishers such as **Titus Software** and **WB Games** have received lower average ratings. Identifying common issues or shortcomings can help us understand the reasons behind lower user satisfaction, such as poor game performance, lack of engaging content, or technical issues.

- **Reviews and Ratings Correlation**: Considering the total number of reviews alongside average ratings allows us to see if there is a correlation between the volume of feedback and overall satisfaction. This helps in identifying whether a high average rating is backed by substantial user feedback or if it is based on a smaller, potentially less reliable sample.

Overall, this analysis not only highlights the current landscape of game publishing but also provides actionable insights that can guide our investigation into potential biases or irregularities in game ratings. By leveraging the strengths of top publishers and learning from the shortcomings of lower-rated ones, we can make data-driven conclusions that enhance the transparency and reliability of game reviews in our study.

## Comparison of User and Journalist Ratings
In this section, we compare the average ratings given by users and journalists. This analysis helps us identify any significant discrepancies or biases in the ratings provided by different groups, providing insights into how professional reviews align or differ from user experiences.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import json

# Load user scores from Parquet file
user_scores_df = pd.read_parquet('data/cleaned_user_scores.parquet')

# Load journalist reviews
with open('data/opencritic_reviews.json', 'r') as file:
    reviews_data = json.load(file)

# Prepare a DataFrame to hold average journalist scores
journalist_scores = []
for game, details in reviews_data.items():
    if 'average_score' in details and isinstance(details['average_score'], (int, float)):
        journalist_scores.append({'Game Name': game, 'Journalist_Score': details['average_score']})

journalist_scores_df = pd.DataFrame(journalist_scores)

# Merge user scores and journalist scores
merged_scores_df = user_scores_df.merge(journalist_scores_df, on='Game Name', how='inner')

# Calculate average user and journalist scores
average_user_score = merged_scores_df['Score'].mean()
average_journalist_score = merged_scores_df['Journalist_Score'].mean()

# Create a bar plot to compare the scores
scores = {'User Score': average_user_score, 'Journalist Score': average_journalist_score}
plt.figure(figsize=(10, 6))
bars = plt.bar(scores.keys(), scores.values(), color=['lightgreen', 'lightblue'])

# Add text labels to each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 2, f'{height:.1f}', ha='center', va='bottom')

plt.xlabel('Source')
plt.ylabel('Average Score')
plt.title('Comparison of Average User and Journalist Scores')
plt.tight_layout()
plt.show()
```

The comparison of average user and journalist ratings reveals a significant discrepancy between the two sources:

- Average User Score: The bar chart shows an average user score of approximately 90. This reflects the collective opinion of the gaming community and their satisfaction with the games.
- Average Journalist Score: The average journalist score is around 97.5, noticeably higher than the user scores. This indicates a more favorable view of the games from professional critics.

The significant difference between user and journalist ratings supports our initial hypothesis that there might be biases in professional reviews. This discrepancy can be attributed to various factors, including potential pressure on journalists to provide positive reviews, access to more polished review copies, or different evaluation criteria used by professionals.

Understanding this difference is crucial for our project as it:

- Highlights Potential Biases: The higher ratings from journalists suggest a potential bias in professional reviews, which could be influenced by factors like marketing relationships or editorial policies.
- Validates User Perspectives: The lower user scores emphasize the importance of incorporating user feedback in our analysis to get a more balanced and realistic view of game quality.
- Informs Review Strategies: By acknowledging this discrepancy, we can develop strategies to critically assess professional reviews and ensure that our evaluations are comprehensive and unbiased.

Overall, this analysis provides valuable insights into the dynamics between user and journalist reviews, reinforcing the need for transparency and accountability in the game review process.

## Correlation Analysis
In this section, we will perform a correlation analysis between normalized user scores, normalized journalist ratings, and various popularity metrics such as the number of reviews, times listed, playing, backlogs, and wishlist. This will help us uncover relationships and patterns within our dataset.

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Function to convert string with 'K' to float
def convert_to_float(value):
    if isinstance(value, str) and 'K' in value:
        return float(value.replace('K', '')) * 1000
    try:
        return float(value)
    except ValueError:
        return None

# Load the games and user scores data
games_df = pd.read_csv('data/games.csv')
user_scores_df = pd.read_parquet('data/cleaned_user_scores.parquet')

# Merge the user scores with game information
merged_df = user_scores_df.merge(games_df, left_on='Game Name', right_on='Title', how='inner')

# Normalize scores to a common scale (0-100)
merged_df['Normalized_User_Score'] = (merged_df['Score'] - merged_df['Score'].min()) / (merged_df['Score'].max() - merged_df['Score'].min()) * 100
merged_df['Normalized_Journalist_Rating'] = (merged_df['Rating'] - merged_df['Rating'].min()) / (merged_df['Rating'].max() - merged_df['Rating'].min()) * 100

# Convert popularity metrics to float
popularity_columns = ['Number of Reviews', 'Times Listed', 'Playing', 'Backlogs', 'Wishlist']
for column in popularity_columns:
    merged_df[column] = merged_df[column].apply(convert_to_float)

# Add popularity metrics to correlation analysis
popularity_corr_df = merged_df[['Normalized_User_Score', 'Normalized_Journalist_Rating', 'Rating'] + popularity_columns]

# Calculate the correlation matrix
popularity_corr_matrix = popularity_corr_df.corr()

# Plot the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(popularity_corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix with Popularity Metrics')
plt.show()
```

### Interpretation

The correlation matrix visualizes the relationships between these variables. Here are some key observations:

- **Normalized User Score and Normalized Journalist Rating**: There is a perfect positive correlation (1.0) between the normalized user scores and normalized journalist ratings, indicating complete consistency between the two metrics when normalized.
- **Number of Reviews and Ratings**: There is a very strong positive correlation (0.88 to 0.9) between the number of reviews and both normalized scores and ratings. This suggests that games with more reviews tend to receive higher or more consistent ratings.
- **Times Listed and Ratings**: A similar strong positive correlation (0.88 to 0.9) exists between times listed and both normalized scores and ratings, indicating that frequently listed games are rated highly.
- **Playing and Ratings**: There is a very strong positive correlation (0.9 to 0.91) between the number of people currently playing the game and both normalized scores and ratings, suggesting that games with high current player counts are highly rated.
- **Backlogs and Ratings**: A slightly lower but still strong positive correlation (0.8 to 0.82) exists between backlogs and both normalized scores and ratings, indicating that games often added to backlogs are also highly rated.
- **Wishlist and Ratings**: There is a very strong positive correlation (0.88 to 0.89) between wishlist counts and both normalized scores and ratings, suggesting that games frequently added to wishlists are also highly rated.

These correlations indicate that popularity metrics such as the number of reviews, times listed, playing, backlogs, and wishlist counts are strongly associated with higher user and journalist ratings. This suggests that more popular games, measured by these metrics, tend to receive higher ratings from both users and journalists.

### Interpretation

The correlation analysis reveals significant relationships that can inform our investigation into potential biases or irregularities in game ratings. For example, the strong correlation between scores and ratings indicates consistency in evaluating game quality. The relationships between the number of reviews, times listed, and plays suggest that popular games with many reviews are also frequently listed and played.

These insights can help us understand how different factors influence game ratings and popularity. By recognizing these patterns, we can better interpret our findings and draw more accurate conclusions about the reliability and transparency of game reviews.

This analysis supports our project's goal of comparing user and journalist reviews to identify discrepancies and biases, ultimately contributing to a more trustworthy gaming review ecosystem.

## Detailed Genre Analysis
In this analysis, we explore if certain genres receive consistently higher or lower ratings from users or journalists. By examining the average ratings for different genres, we can gain insights into user preferences and journalist evaluations across various game categories.

The chart illustrates the differences in average normalized ratings by genre:
- User Score: Represents the user scores gathered from user review platforms like Backloggd.
- Journalist Rating: Represents the ratings given by journalists collected from OpenCritic.

```python
# Load the necessary data
import pandas as pd
import matplotlib.pyplot as plt

# Load the games and user scores data
games_df = pd.read_csv('data/games.csv')
user_scores_df = pd.read_parquet('data/cleaned_user_scores.parquet')

# Merge the user scores with game information
merged_df = user_scores_df.merge(games_df, left_on='Game Name', right_on='Title', how='inner')

# Use the correct genre column
genres_column = 'Genres_x' if 'Genres_x' in merged_df.columns else 'Genres_y'

# Extract genre information and calculate average ratings by genre
merged_df[genres_column] = merged_df[genres_column].apply(lambda x: x.strip("[]").replace("'", "").split(", ") if pd.notnull(x) else [])

# Normalize scores to a common scale (0-100)
merged_df['Normalized_User_Score'] = (merged_df['Score'] - merged_df['Score'].min()) / (merged_df['Score'].max() - merged_df['Score'].min()) * 100
merged_df['Normalized_Journalist_Rating'] = (merged_df['Rating'] - merged_df['Rating'].min()) / (merged_df['Rating'].max() - merged_df['Rating'].min()) * 100

# Calculate average normalized ratings by genre
genre_ratings = merged_df.explode(genres_column).groupby(genres_column)[['Normalized_User_Score', 'Normalized_Journalist_Rating']].mean().reset_index()
genre_ratings = genre_ratings.dropna()

# Plot average user and journalist ratings by genre
plt.figure(figsize=(14, 8))
genre_ratings.plot(kind='bar', x=genres_column, y=['Normalized_User_Score', 'Normalized_Journalist_Rating'], color=['lightgreen', 'lightblue'])
plt.xlabel('Genre')
plt.ylabel('Average Normalized Rating')
plt.title('Average Normalized Ratings by Genre')
plt.legend(['User Score', 'Journalist Rating'])
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

The chart compares average normalized ratings by genre for user scores and journalist ratings:

- Adventure: Both users and journalists rate adventure games similarly, around 70.
- RPG: RPGs receive high ratings from both groups, close to 100.
- Shooter: Shooter games have lower ratings, around 20-30, indicating less preference from both groups.
- Simulator: Simulator games are rated moderately, around 50, with little difference between users and journalists.
- Sport: Sports games also receive moderate ratings, around 50-60.

This analysis shows general agreement between users and journalists, especially for RPGs and adventure games, with both groups showing less enthusiasm for shooter games.

## Average Ratings Over Time
This analysis will explore how average ratings for games have evolved over time, comparing user and journalist ratings across different years. This can help us understand trends and shifts in game quality perception over the years.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the games data
games_df = pd.read_csv('data/games.csv')

# Convert the 'Release Date' column to datetime
games_df['Release Date'] = pd.to_datetime(games_df['Release Date'], errors='coerce')

# Extract the year and quarter from the release date
games_df['YearQuarter'] = games_df['Release Date'].dt.to_period('Q')

# Calculate the average rating by year and quarter
avg_ratings_by_quarter = games_df.groupby('YearQuarter')['Rating'].mean().reset_index()

# Plot the average ratings over time with a trendline
plt.figure(figsize=(14, 8))
sns.lineplot(x=avg_ratings_by_quarter['YearQuarter'].astype(str), y=avg_ratings_by_quarter['Rating'], label='Average Rating', color='blue')
sns.regplot(x=avg_ratings_by_quarter.index, y=avg_ratings_by_quarter['Rating'], scatter=False, label='Trendline', color='red')
plt.xlabel('Year-Quarter')
plt.ylabel('Average Rating')
plt.title('Average Game Ratings Over Time')
plt.xticks(ticks=range(0, len(avg_ratings_by_quarter), 8), labels=avg_ratings_by_quarter['YearQuarter'].astype(str)[::8], rotation=45, fontsize=8)
plt.legend()
plt.tight_layout()
plt.show()
```

The plot illustrates the average game ratings over time, aggregated by quarterly periods, with a trendline indicating the overall direction of the ratings. The trendline suggests a slight increase in average ratings over time, indicating a general improvement in game quality or reception. This trend can provide insights into how the gaming industry has evolved and help identify periods of particularly high or low game quality. This analysis aligns with our project objective of understanding the dynamics between user and journalist reviews and identifying potential biases or irregularities in game ratings.

## Average Ratings by Media Outlet

In this section, we analyze the average ratings given by different media outlets. This helps us understand if certain outlets tend to give higher or lower ratings compared to others.

### Methodology
- We extracted ratings from OpenCritic reviews and calculated the average score for each media outlet.
- The data was aggregated and visualized in a bar chart to compare average ratings across different outlets.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json

# Load the reviews data
with open('data/opencritic_reviews.json', 'r') as file:
    reviews_data = json.load(file)

# Prepare a DataFrame from the JSON data
media_ratings = []
for game, details in reviews_data.items():
    outlets = details.get('outlets', {})
    for outlet, review in outlets.items():
        score = review.get('score')
        if isinstance(score, (int, float)):
            media_ratings.append({
                'Game Name': game,
                'Media Outlet': outlet,
                'Score': score
            })

media_ratings_df = pd.DataFrame(media_ratings)

# Calculate average rating by media outlet
avg_ratings_by_media = media_ratings_df.groupby('Media Outlet')['Score'].mean().reset_index()

# Sort by average rating in descending order
avg_ratings_by_media = avg_ratings_by_media.sort_values(by='Score', ascending=False)

# Plot the average ratings by media outlet
plt.figure(figsize=(14, 8))
bars = sns.barplot(x='Score', y='Media Outlet', data=avg_ratings_by_media, palette='coolwarm', dodge=False)
plt.xlabel('Average Rating')
plt.ylabel('Media Outlet')
plt.title('Average Ratings by Media Outlet')

# Add text labels to each bar
for bar in bars.patches:
    plt.text(bar.get_width() + 0.1, bar.get_y() + bar.get_height() / 2, f'{bar.get_width():.1f}', ha='left', va='center')

plt.tight_layout()
plt.show()
```

The bar chart displays the average ratings given by different media outlets. This analysis helps us understand if certain media outlets tend to give higher or lower ratings compared to others. By examining these differences, we can identify potential biases or preferences that may exist within specific outlets. This information is crucial for understanding how media outlet ratings align with user ratings and for ensuring transparency and accountability in the review process.

## Average Ratings by Author

In this section, we analyze the average ratings given by different authors. This analysis helps us understand if certain authors tend to give consistently higher or lower ratings compared to others. By examining the average ratings, we can gain insights into potential biases or preferences that individual authors might have.

### Methodology
- We extracted ratings from OpenCritic reviews and calculated the average score for each author.
- The data was aggregated and visualized in a bar chart to compare average ratings across different authors.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json

# Load the reviews data
with open('data/opencritic_reviews.json', 'r') as file:
    reviews_data = json.load(file)

# Prepare a list to hold author and their scores
author_ratings = []

# Extract author names and their scores from the reviews data
for game, details in reviews_data.items():
    if 'outlets' in details:
        for outlet, outlet_data in details['outlets'].items():
            if 'author' in outlet_data and 'score' in outlet_data:
                author_name = outlet_data['author']
                score = outlet_data['score']
                if isinstance(score, (int, float)):
                    author_ratings.append({'Author': author_name, 'Score': score})

# Create a DataFrame from the author ratings list
author_ratings_df = pd.DataFrame(author_ratings)

# Calculate average ratings by author
avg_ratings_by_author = author_ratings_df.groupby('Author')['Score'].mean().reset_index()

# Sort the DataFrame by score in descending order
avg_ratings_by_author = avg_ratings_by_author.sort_values(by='Score', ascending=False)

# Plot average ratings by author
plt.figure(figsize=(14, 8))
bars = sns.barplot(x='Score', y='Author', data=avg_ratings_by_author, hue='Author', dodge=False, palette='coolwarm')

# Add labels to the bars
for bar in bars.patches:
    bars.annotate(format(bar.get_width(), '.1f'),
                  (bar.get_width(), bar.get_y() + bar.get_height() / 2),
                  ha='center', va='center',
                  size=9, xytext=(40, 0),
                  textcoords='offset points')

plt.xlabel('Average Rating')
plt.ylabel('Author')
plt.title('Average Ratings by Author')
plt.legend([],[], frameon=False)
plt.tight_layout()
plt.show()
```

The bar chart above shows the average ratings given by different authors. The ratings are almost uniformly high, with most authors giving an average rating of 100.0. This might suggest a lack of variability or a potential bias towards higher ratings among the authors listed. Notably, only two authors, Brad Ellis and Tyler Colp, have given average ratings below 100.0, with averages of 95.0 and 90.0, respectively. This observation could indicate that these authors are more critical in their reviews compared to their peers. However, the overall trend of high average ratings raises questions about the consistency and objectivity of the reviews across different authors.

```python
import json
import requests

# Connect to rapid api's opencritic API
url = "https://opencritic-api.p.rapidapi.com/review/author/481?skip=20&sort=newest"
headers = {
    "X-RapidAPI-Host": "opencritic-api.p.rapidapi.com",
    "X-RapidAPI-Key": "64686b9a75msha576b74f8ff0b56p1a63ebjsnf20f69b4fac9"
}

response = requests.get(url, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    data = response.json()

    file_path = "opencritic_reviews.json"

    # Open the file in write mode and save the JSON data
    with open(file_path, "w") as file:
        json.dump(data, file)

    print("JSON data saved to:", file_path)
else:
    print("Error:", response.status_code)
```
